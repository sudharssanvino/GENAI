
Stack - LIFO - Browsing history

```python
browser_history = ['google.com', 'stackoverflow.com', 'github.com']
print(browser_history)
```

```python
browser_history.append('python.org')
print(browser_history)
```

```python
last_page = browser_history.pop()
print(f"Last visited page: {last_page}")
print(browser_history)
```

Queue - FIFO

```python
from collections import deque
ticket_queue = deque()
ticket_queue.append('Ticket 1') ## Enqueue
ticket_queue.append('Ticket 2')
print(ticket_queue)
```

```python
served = ticket_queue.popleft() ## Deque
print(f"Serving: {served}")
```

FrozenSet - Immutable datatype

```python
phonenumber = {frozenset({'+91'}) : '9876543210'}
print(phonenumber)
```

```python
from collections import namedtuple
```
