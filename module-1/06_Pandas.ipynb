## Example 1: Creating Series and DataFrames

The two building blocks of Pandas.

```python
import pandas as pd
scores = pd.Series([45,58,84,62], index=["Alice", "Bob", "Charlie", "David"], name="Scores")
print(scores)

```

```python
import pandas as pd
import numpy as np

# --- Series: a single labeled column ---
scores = pd.Series([88, 92, 75, 95, 60],
                   index=['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
                   name='Exam Score')
print("Series:")
print(scores)
print("\nAccess Bob's score:", scores['Bob'])
print("All scores above 80:\n", scores[scores > 80])

# --- DataFrame: a full table ---
data = {
    'Name'      : ['Alice', 'Bob', 'Carol', 'Dave', 'Eve'],
    'Age'       : [24, 30, 27, 35, 22],
    'Department': ['Engineering', 'Marketing', 'Engineering', 'HR', 'Marketing'],
    'Salary'    : [72000, 55000, 80000, 62000, 48000],
    'Experience': [2, 6, 4, 10, 1],
}
df = pd.DataFrame(data)
print("\nDataFrame:")
print(df)
```

## Example 2: Reading and Inspecting Data

The first thing you do with any new dataset — understand its structure.

```python
import pandas as pd
import numpy as np

# Create a realistic employee dataset
np.random.seed(42)
n = 50
df = pd.DataFrame({
    'emp_id'    : range(1001, 1001+n),
    'name'      : [f'Employee_{i}' for i in range(n)],
    'age'       : np.random.randint(22, 60, n),
    'dept'      : np.random.choice(['Engineering','Marketing','HR','Finance','Sales'], n),
    'salary'    : np.random.randint(35000, 120000, n),
    'years_exp' : np.random.randint(0, 25, n),
    'rating'    : np.random.choice([1,2,3,4,5], n),
})
# Add some missing values intentionally
df.loc[np.random.choice(df.index, 5), 'salary'] = np.nan

print("--- shape ---")
print(df.shape)                 # (rows, columns)

print("\n--- first 5 rows ---")
print(df.head())

print("\n--- last 3 rows ---")
print(df.tail(3))

print("\n--- column names ---")
print(df.columns.tolist())

print("\n--- dtypes ---")
print(df.dtypes)

print("\n--- describe (statistics) ---")
print(df.describe())

print("\n--- info (null counts + dtypes) ---")
df.info()
```

## Example 3: Selecting Rows and Columns

Accessing specific data using `.loc`, `.iloc`, and column names.

```python
import pandas as pd

data = {
    'Name'  : ['Alice','Bob','Carol','Dave','Eve','Frank'],
    'Dept'  : ['Eng','Mkt','Eng','HR','Mkt','Eng'],
    'Salary': [72000,55000,80000,62000,48000,91000],
    'Age'   : [24,30,27,35,22,40],
    'Rating': [4,3,5,3,2,5],
}
df = pd.DataFrame(data)
print("Full DataFrame:\n", df)

# Select single column (returns Series)
print("\nSalary column:\n", df['Salary'].tolist())

# Select multiple columns (returns DataFrame)
print("\nName + Salary:\n", df[['Name','Salary']])

# .loc  — label-based selection [row_label, col_label]
print("\nloc row 2  :", df.loc[2].to_dict())
print("loc rows 1-3:\n", df.loc[1:3, ['Name','Salary']])

# .iloc — position-based selection [row_pos, col_pos]
print("\niloc row 0 :", df.iloc[0].to_dict())
print("iloc last 2 rows, first 3 cols:\n", df.iloc[-2:, :3])
```

## Example 4: Filtering Data

Selecting rows that meet specific conditions — the most common daily operation.

```python
import pandas as pd

data = {
    'Name'  : ['Alice','Bob','Carol','Dave','Eve','Frank','Grace'],
    'Dept'  : ['Eng','Mkt','Eng','HR','Mkt','Eng','HR'],
    'Salary': [72000,55000,80000,62000,48000,91000,67000],
    'Age'   : [24,30,27,35,22,40,29],
    'Rating': [4,3,5,3,2,5,4],
}
df = pd.DataFrame(data)

# Single condition
high_salary = df[df['Salary'] > 65000]
print("Salary > 65,000:\n", high_salary, "\n")

# Multiple conditions with & (AND) and | (OR)
eng_senior = df[(df['Dept'] == 'Eng') & (df['Age'] > 25)]
print("Engineering AND Age > 25:\n", eng_senior, "\n")

top_or_exp = df[(df['Rating'] == 5) | (df['Age'] > 35)]
print("Rating=5 OR Age>35:\n", top_or_exp, "\n")

# .isin() — filter by a list of values
mkt_hr = df[df['Dept'].isin(['Mkt','HR'])]
print("Dept in [Mkt, HR]:\n", mkt_hr, "\n")

# .between() — range filter
mid_salary = df[df['Salary'].between(60000, 80000)]
print("Salary between 60k-80k:\n", mid_salary)
```

## Example 5: Adding, Renaming, and Dropping Columns

Transforming your DataFrame structure.

```python
import pandas as pd

data = {
    'Name'  : ['Alice','Bob','Carol','Dave','Eve'],
    'Salary': [72000,55000,80000,62000,48000],
    'Age'   : [24,30,27,35,22],
    'Rating': [4,3,5,3,2],
}
df = pd.DataFrame(data)
print("Original:\n", df)

# Add new column: derived from existing
df['Annual_Bonus']   = df['Salary'] * 0.10
df['Total_Package']  = df['Salary'] + df['Annual_Bonus']
df['Senior']         = df['Age'] > 28       # boolean column
df['Grade']          = df['Rating'].map({1:'F',2:'D',3:'C',4:'B',5:'A'})

print("\nAfter adding columns:\n", df)

# Rename columns
df = df.rename(columns={'Salary':'Base_Salary', 'Age':'Employee_Age'})
print("\nAfter rename:\n", df.columns.tolist())

# Drop columns
df = df.drop(columns=['Annual_Bonus'])
print("After dropping Annual_Bonus:", df.columns.tolist())

# Drop rows
df = df.drop(index=[0, 4])   # drop Alice and Eve
print("\nAfter dropping rows 0 and 4:\n", df.reset_index(drop=True))
```

## Example 6: Handling Missing Values

Real datasets always have missing data. This is how to find, fill, and remove it.

```python
import pandas as pd
import numpy as np

# Dataset with intentional missing values
data = {
    'Name'  : ['Alice','Bob','Carol','Dave','Eve','Frank'],
    'Age'   : [24, np.nan, 27, 35, np.nan, 40],
    'Salary': [72000, 55000, np.nan, 62000, 48000, np.nan],
    'Dept'  : ['Eng', 'Mkt', 'Eng', None, 'Mkt', 'Eng'],
    'Score' : [85, 90, np.nan, 78, 82, np.nan],
}
df = pd.DataFrame(data)
print("Dataset with missing values:\n", df)

# Detect missing values
print("\n--- isna() map ---")
print(df.isna())

print("\n--- Missing count per column ---")
print(df.isna().sum())

print("\n--- Percentage missing ---")
print((df.isna().sum() / len(df) * 100).round(1))

# Drop rows with ANY missing value
df_dropped = df.dropna()
print("\nAfter dropna():", df_dropped.shape, "rows remain")

# Fill missing values
df_filled = df.copy()
df_filled['Age']    = df_filled['Age'].fillna(df_filled['Age'].median())
df_filled['Salary'] = df_filled['Salary'].fillna(df_filled['Salary'].mean())
df_filled['Dept']   = df_filled['Dept'].fillna('Unknown')
print("\nAfter filling missing values:\n", df_filled)
```

## Example 7: GroupBy and Aggregation

The Pandas equivalent of SQL GROUP BY — powerful for summarising data.

```python
import pandas as pd
import numpy as np

np.random.seed(0)
df = pd.DataFrame({
    'Dept'    : np.random.choice(['Engineering','Marketing','HR','Finance'], 40),
    'Gender'  : np.random.choice(['Male','Female'], 40),
    'Salary'  : np.random.randint(40000, 120000, 40),
    'Rating'  : np.random.randint(1, 6, 40),
    'Years'   : np.random.randint(0, 20, 40),
})

# Basic groupby — one column
print("--- Average salary by department ---")
print(df.groupby('Dept')['Salary'].mean().round(0))

# Multiple aggregations with agg()
print("\n--- Dept summary (count, mean salary, max rating) ---")
summary = df.groupby('Dept').agg(
    Count      = ('Salary', 'count'),
    Avg_Salary = ('Salary', 'mean'),
    Max_Rating = ('Rating', 'max'),
    Avg_Years  = ('Years',  'mean'),
).round(1)
print(summary)

# Group by multiple columns
print("\n--- Avg salary by Dept + Gender ---")
multi = df.groupby(['Dept','Gender'])['Salary'].mean().round(0).unstack()
print(multi)

# value_counts — frequency of categories
print("\n--- Employee count per department ---")
print(df['Dept'].value_counts())
```

## Example 8: Sorting and Ranking

Ordering data by values — useful for leaderboards, top-N lists, and reports.

```python
import pandas as pd

data = {
    'Product' : ['Laptop','Phone','Tablet','Monitor','Keyboard','Mouse','Headset'],
    'Category': ['Tech','Tech','Tech','Tech','Accessory','Accessory','Accessory'],
    'Sales'   : [4200, 8900, 3100, 2700, 5600, 6800, 1900],
    'Revenue' : [630000, 445000, 217000, 189000, 112000, 68000, 47500],
    'Rating'  : [4.5, 4.8, 4.2, 4.0, 4.6, 4.7, 4.1],
}
df = pd.DataFrame(data)
print("Product data:\n", df)

# Sort by single column
print("\n--- Sorted by Revenue (highest first) ---")
print(df.sort_values('Revenue', ascending=False)[['Product','Revenue']])

# Sort by multiple columns
print("\n--- Sorted by Category, then Rating (desc) ---")
print(df.sort_values(['Category','Rating'], ascending=[True, False])[['Product','Category','Rating']])

# Get top-3 products by sales
print("\n--- Top 3 by Sales ---")
print(df.nlargest(3, 'Sales')[['Product','Sales']])

# Get bottom-2 by revenue
print("\n--- Bottom 2 by Revenue ---")
print(df.nsmallest(2, 'Revenue')[['Product','Revenue']])
```

## Example 9: Merging DataFrames

Combining two tables — equivalent to SQL JOINs.

```python
import pandas as pd

# Table 1: Employee info
employees = pd.DataFrame({
    'emp_id'  : [101, 102, 103, 104, 105],
    'name'    : ['Alice','Bob','Carol','Dave','Eve'],
    'dept_id' : [10, 20, 10, 30, 20],
})

# Table 2: Department info
departments = pd.DataFrame({
    'dept_id'  : [10, 20, 30, 40],
    'dept_name': ['Engineering','Marketing','HR','Finance'],
    'location' : ['NYC','LA','Chicago','Boston'],
})

# Table 3: Salaries (missing employee 105)
salaries = pd.DataFrame({
    'emp_id'  : [101, 102, 103, 104],
    'salary'  : [72000, 55000, 80000, 62000],
})

print("Employees:\n", employees)
print("\nDepartments:\n", departments)
print("\nSalaries:\n", salaries)

# INNER JOIN — only matching rows
inner = pd.merge(employees, departments, on='dept_id', how='inner')
print("\n--- INNER JOIN (employees + departments) ---")
print(inner)

# LEFT JOIN — all employees, even if no salary
left = pd.merge(employees, salaries, on='emp_id', how='left')
print("\n--- LEFT JOIN (all employees + salary where available) ---")
print(left)

# Chain merges
full = pd.merge(employees, departments, on='dept_id')          .merge(salaries, on='emp_id', how='left')
print("\n--- Full merged table ---")
print(full)

# Concatenate — stack DataFrames vertically
q1 = pd.DataFrame({'quarter':['Q1'],'sales':[1200]})
q2 = pd.DataFrame({'quarter':['Q2'],'sales':[1500]})
combined = pd.concat([q1, q2], ignore_index=True)
print("\n--- Concatenated ---")
print(combined)
```

## Example 10: String Operations

Cleaning and transforming text data using `.str` accessor.

```python
import pandas as pd

data = {
    'full_name' : ['  alice Johnson ', 'BOB smith', 'carol WANG', 'Dave OBrien', '  Eve KUMAR  '],
    'email'     : ['alice@company.com','bob@gmail.com','carol@company.com','dave@outlook.com','eve@company.com'],
    'phone'     : ['(123) 456-7890','456-789-0123','(789) 012-3456','012-345-6789','(345) 678-9012'],
    'dept_code' : ['ENG-001','MKT-002','ENG-003','HR-004','MKT-005'],
}
df = pd.DataFrame(data)
print("Raw data:\n", df)

# Clean names
df['clean_name']  = df['full_name'].str.strip().str.title()
df['first_name']  = df['clean_name'].str.split().str[0]
df['last_name']   = df['clean_name'].str.split().str[-1]

# Email operations
df['email_domain'] = df['email'].str.split('@').str[1]
df['is_company']   = df['email'].str.endswith('@company.com')

# String checks
df['dept']         = df['dept_code'].str[:3]            # first 3 chars
df['dept_num']     = df['dept_code'].str.extract(r'(\d+)')  # extract digits

# Replace / remove characters
df['phone_clean']  = df['phone'].str.replace(r'[^\d]','', regex=True)  # digits only

print("\nCleaned data:")
print(df[['clean_name','first_name','email_domain','is_company','dept','phone_clean']])

# String filtering
company_staff = df[df['is_company'] == True]
print("\nCompany email employees:", company_staff['first_name'].tolist())

eng_team = df[df['dept'] == 'ENG']
print("Engineering team:", eng_team['first_name'].tolist())
```
