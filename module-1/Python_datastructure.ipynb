```python
import pandas as pd
import numpy as np
 
np.random.seed(42)
n = 50
emp_data = {
    "EmployeeID": np.arange(1000, n + 1000),
    "Name": [f"Employee_{i}" for i in range(1, n + 1)],
    "Department": np.random.choice(["HR", "Finance", "IT", "Marketing"], size=n),
    "Age": np.random.randint(22, 60, size=n),
    "Salary": np.random.randint(40000, 120000, size=n)
}
 
df = pd.DataFrame(emp_data)
print(df.head())
print(df.dtypes)
print(df.describe())
print(df.info())
 
print("Salary",df[["Name","Salary"]])
 
print("Locate Row 25:", df.loc[25].to_dict())
 
print("Locate Row 25:", df.loc[20:25][["Name","Salary"]])
```

```python
import pandas as pd
import numpy as np
 
# Dataset with intentional missing values
data = {
    'Name'  : ['Alice','Bob','Carol','Dave','Eve','Frank'],
    'Age'   : [24, np.nan, 27, 35, np.nan, 40],
    'Salary': [72000, 55000, np.nan, 62000, 48000, np.nan],
    'Dept'  : ['Eng', 'Mkt', 'Eng', None, 'Mkt', 'Eng'],
    'Score' : [85, 90, np.nan, 78, 82, np.nan],
}
df = pd.DataFrame(data)
 
print("Original DataFrame with Missing Values:")
print(df)
print("Missing Values:")
print(df.isna())
 
df_filled = df.copy()
df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].mean())
print("\nDataFrame after filling missing values in 'Age' with mean:")
print(df_filled)
# df_dropped = df.dropna()
# print("DataFrame after dropping missing values:")
# print(df_dropped)
print("\nDataFrame sorted by Salary in descending order:")
print(df.sort_values(by='Salary', ascending=False ))
 
print("\nTop 3 Employees by Salary:")
print(df.nlargest(2, 'Salary')) 
```

DOG Image working using NUMPY

```python
import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

image_path = "dog.jpg"

if "image" in globals():
	# reuse already loaded image if present in notebook state
	image = globals()["image"]
elif os.path.exists(image_path):
	image = Image.open(image_path)
else:
	raise FileNotFoundError(
		f"'{image_path}' not found. Put the file into the current working directory "
		"or update image_path to the correct path."
	)

image_array = np.array(image)
plt.imshow(image_array)
plt.axis("off")
plt.show()
```
