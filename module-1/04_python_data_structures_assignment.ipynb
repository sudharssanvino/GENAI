## 1️⃣ LIST

### 📌 What is it?
A **List** is an ordered, mutable (changeable) collection that can hold items of any type. Items are stored in sequence and can be accessed by index.

### 🌍 Real-world analogy
Think of a **shopping cart** — you can add items, remove them, or reorder them anytime.  
Example: Your grocery list: `['milk', 'eggs', 'bread']` — order matters, and you can edit it.

### ✅ Why choose it?
- You need an **ordered** collection
- You need to **add, remove, or update** items frequently
- You want to allow **duplicate values**
- Best for: to-do lists, queues, dynamic collections

````python
# 🛒 Shopping Cart — List Example
cart = ['milk', 'eggs', 'bread']

cart.append('butter')       # Add item
cart.remove('eggs')         # Remove item
cart[0] = 'oat milk'        # Update item

print('Cart items:', cart)
print('First item:', cart[0])
print('Total items:', len(cart))
````

## 2️⃣ TUPLE

### 📌 What is it?
A **Tuple** is an ordered, **immutable** (unchangeable) collection. Once created, its items cannot be modified.

### 🌍 Real-world analogy
Think of **GPS coordinates** — `(13.0827, 80.2707)` for Chennai. The latitude and longitude are fixed; you'd never accidentally change them.  
Another example: A date of birth `(1995, 6, 15)` — it never changes.

### ✅ Why choose it?
- You have **fixed data** that should never change
- Tuples are **faster** than lists
- Can be used as **dictionary keys** (lists cannot)
- Best for: coordinates, RGB colors, DB records, constants

````python
# 📍 GPS Coordinates — Tuple Example
chennai_coords = (13.0827, 80.2707)   # (latitude, longitude) — fixed forever
rgb_red = (255, 0, 0)                 # Red color — never changes

print('Chennai coordinates:', chennai_coords)
print('Latitude:', chennai_coords[0])
print('Red color RGB:', rgb_red)

# Unpacking a tuple
lat, lon = chennai_coords
print(f'Lat={lat}, Lon={lon}')

# Trying to change will raise an error:
# chennai_coords[0] = 99  ❌ TypeError: 'tuple' object does not support item assignment
````

## 3️⃣ SET

### 📌 What is it?
A **Set** is an unordered collection of **unique** items. Duplicates are automatically removed. Supports math operations like union, intersection.

### 🌍 Real-world analogy
Think of **tags on a blog post** — you can't tag the same post with 'python' twice. Duplicates simply don't exist.  
Or a **guest list** — each person appears only once.

### ✅ Why choose it?
- You need to store **unique values only**
- You want **super-fast membership checks** (`in` operator)
- You need **set operations**: union, intersection, difference
- Best for: removing duplicates, finding common items, tag systems

````python
# 🏷️ Blog Tags — Set Example
post1_tags = {'python', 'coding', 'tutorial', 'python'}  # duplicate 'python' ignored
post2_tags = {'coding', 'data', 'tutorial'}

print('Post 1 tags:', post1_tags)           # No duplicate 'python'
print('Common tags:', post1_tags & post2_tags)  # Intersection
print('All tags:', post1_tags | post2_tags)     # Union
print('Only in Post1:', post1_tags - post2_tags) # Difference

# Super-fast lookup
print('Has python tag?', 'python' in post1_tags)
````

## 4️⃣ DICTIONARY (dict)

### 📌 What is it?
A **Dictionary** stores data as **key-value pairs**. You look up values using unique keys, just like a real dictionary where you look up a word to get its meaning.

### 🌍 Real-world analogy
Think of a **student report card** — each subject (key) maps to a score (value).  
`{'Maths': 95, 'Science': 88, 'English': 76}`

### ✅ Why choose it?
- You need to **look up data by a name/label** (not position)
- You want to store **structured records** (like JSON)
- **O(1) average lookup** — extremely fast
- Best for: configs, user profiles, counting frequency, JSON data

````python
# 📋 Student Report Card — Dictionary Example
report_card = {
    'name': 'Arjun',
    'Maths': 95,
    'Science': 88,
    'English': 76
}

print('Student:', report_card['name'])
print('Maths score:', report_card['Maths'])

# Add a new subject
report_card['Tamil'] = 92

# Calculate average
subjects = ['Maths', 'Science', 'English', 'Tamil']
avg = sum(report_card[s] for s in subjects) / len(subjects)
print(f'Average score: {avg:.1f}')
print('All subjects:', list(report_card.keys()))
````

## 5️⃣ ARRAY (from `array` module)

### 📌 What is it?
An **Array** is like a list but stores elements of **only one data type** (e.g., all integers or all floats). More memory-efficient than lists for numeric data.

### 🌍 Real-world analogy
Think of **daily temperature readings** — all are numbers, nothing else. Storing 10,000 temperature values as an array uses far less memory than a list.

### ✅ Why choose it?
- You are storing **large amounts of numeric data**
- You need **memory efficiency** (no mixed types)
- You're doing **low-level numeric processing**
- Best for: sensor readings, audio samples, binary data buffers
> 💡 For heavy math, prefer **NumPy arrays** instead.

````python
# 🌡️ Daily Temperatures — Array Example
import array

# 'f' = float type code
temperatures = array.array('f', [28.5, 30.1, 27.8, 32.0, 29.4])

print('Temperatures:', list(temperatures))
print('Hottest day:', max(temperatures))
print('Average temp:', round(sum(temperatures) / len(temperatures), 2))

# Add a new reading
temperatures.append(31.5)
print('After new reading:', list(temperatures))

# Memory comparison
import sys
py_list = [28.5, 30.1, 27.8, 32.0, 29.4]
print(f'List size: {sys.getsizeof(py_list)} bytes')
print(f'Array size: {sys.getsizeof(temperatures)} bytes  ← more compact!')
````

## 6️⃣ STRING (str)

### 📌 What is it?
A **String** is an immutable sequence of characters. It's a data structure too — you can index, slice, and iterate over it just like a list.

### 🌍 Real-world analogy
Think of a **train ticket** — your name, PNR number, seat — all text. You read it, but you can't erase letters off a printed ticket (immutable).

### ✅ Why choose it?
- Storing and processing **text data**
- Rich built-in methods: `split`, `join`, `replace`, `upper`, etc.
- Best for: names, messages, file paths, templates, parsing

````python
# 🎫 Train Ticket Info — String Example
ticket = "PNR:1234567890 | NAME:ARJUN KUMAR | SEAT:A1 | FROM:MAS | TO:NDLS"

print('Ticket:', ticket)
print('Uppercase:', ticket.upper())

# Extract parts
parts = ticket.split(' | ')
for part in parts:
    key, value = part.split(':')
    print(f'  {key.strip()} → {value.strip()}')

# Check and slice
print('Has Arjun?', 'ARJUN' in ticket)
print('First 3 chars:', ticket[:3])
````

## 7️⃣ STACK (using list)

### 📌 What is it?
A **Stack** follows **LIFO** — Last In, First Out. The last item added is the first to be removed. Python implements this using a list with `append()` and `pop()`.

### 🌍 Real-world analogy
Think of **browser history** — the last page you visited is the first one you go back to when you press the Back button.

### ✅ Why choose it?
- Undo/Redo functionality
- Function call management (call stack)
- Backtracking in algorithms
- Best for: browser history, expression evaluation, DFS

````python
# 🌐 Browser History — Stack Example
browser_history = []

# Visit pages
browser_history.append('google.com')
browser_history.append('youtube.com')
browser_history.append('github.com')

print('History:', browser_history)
print('Current page:', browser_history[-1])

# Press Back button
last_page = browser_history.pop()
print(f'Left: {last_page}')
print(f'Now on: {browser_history[-1]}')
print('Remaining history:', browser_history)
````

## 8️⃣ QUEUE (using deque)

### 📌 What is it?
A **Queue** follows **FIFO** — First In, First Out. The first item added is the first to leave. Python's `collections.deque` is the efficient way to implement this.

### 🌍 Real-world analogy
Think of a **ticket counter queue** — the person who arrives first gets served first. No cutting in line!

### ✅ Why choose it?
- Task scheduling (print queue, CPU tasks)
- BFS (Breadth-First Search) in graphs
- Handling requests in order
- Best for: customer service systems, order processing, BFS

````python
# 🎟️ Ticket Counter Queue — Queue Example
from collections import deque

ticket_queue = deque()

# People join the queue
ticket_queue.append('Arjun')
ticket_queue.append('Priya')
ticket_queue.append('Ravi')

print('Queue:', list(ticket_queue))

# Serve customers one by one
served = ticket_queue.popleft()   # First person out
print(f'Now serving: {served}')
print(f'Remaining queue: {list(ticket_queue)}')

# New person joins
ticket_queue.append('Meena')
print(f'Updated queue: {list(ticket_queue)}')
````

## 9️⃣ FROZENSET

### 📌 What is it?
A **Frozenset** is an **immutable version of a set**. Like a set (unique items, unordered), but it cannot be changed after creation.

### 🌍 Real-world analogy
Think of **country calling codes** — `{+91, +1, +44}`. These are a fixed set of unique values that don't change. You'd use a frozenset as a dictionary key (sets can't be used as keys).

### ✅ Why choose it?
- Need set features but want **immutability**
- Use as **dictionary keys** or inside another set
- Best for: fixed categories, permission groups, hashable sets

````python
# 🔐 Permission Roles — Frozenset Example
admin_permissions = frozenset({'read', 'write', 'delete', 'execute'})
user_permissions  = frozenset({'read', 'write'})

print('Admin permissions:', admin_permissions)
print('User permissions:', user_permissions)
print('User is subset of Admin?', user_permissions.issubset(admin_permissions))
print('Extra admin perms:', admin_permissions - user_permissions)

# Use as dict key (regular set can't do this!)
role_map = {
    frozenset({'read'}): 'Viewer',
    frozenset({'read', 'write'}): 'Editor',
    frozenset({'read', 'write', 'delete', 'execute'}): 'Admin'
}
print('Role for user:', role_map[user_permissions])
````

## 🔟 NAMED TUPLE

### 📌 What is it?
A **Named Tuple** is a tuple where each position has a **name**. Combines the immutability of tuples with the readability of dictionaries — without the overhead.

### 🌍 Real-world analogy
Think of an **employee badge** — it has fixed fields: name, ID, department. You access them by name (`badge.name`) not by index (`badge[0]`), making code readable.

### ✅ Why choose it?
- Tuples where field names make code **self-documenting**
- Lighter than a full class or dict
- Best for: records, CSV rows, coordinates with labels, config structs

````python
# 🪪 Employee Badge — Named Tuple Example
from collections import namedtuple

Employee = namedtuple('Employee', ['name', 'emp_id', 'department', 'salary'])

emp1 = Employee(name='Priya', emp_id='E101', department='Engineering', salary=85000)
emp2 = Employee(name='Ravi',  emp_id='E102', department='HR',          salary=60000)

print(f'Name: {emp1.name}, Dept: {emp1.department}')  # Access by name
print(f'Salary: ₹{emp1.salary:,}')

# Still works like a tuple
print('As tuple:', tuple(emp2))
print('Dept by index:', emp1[2])  # same as emp1.department
````

## 📊 Quick Comparison Cheat Sheet

| Data Structure | Ordered | Mutable | Duplicates | Key-Value | Best Use Case |
|---|---|---|---|---|---|
| **List** | ✅ | ✅ | ✅ | ❌ | Dynamic collections, sequences |
| **Tuple** | ✅ | ❌ | ✅ | ❌ | Fixed records, coordinates |
| **Set** | ❌ | ✅ | ❌ | ❌ | Unique items, fast lookup |
| **Dict** | ✅* | ✅ | Keys: ❌ | ✅ | Key-value lookup, structured data |
| **Array** | ✅ | ✅ | ✅ | ❌ | Numeric data, memory efficiency |
| **String** | ✅ | ❌ | ✅ | ❌ | Text processing |
| **Stack** | ✅ | ✅ | ✅ | ❌ | LIFO — undo, history |
| **Queue** | ✅ | ✅ | ✅ | ❌ | FIFO — scheduling, BFS |
| **Frozenset** | ❌ | ❌ | ❌ | ❌ | Immutable unique items, dict keys |
| **NamedTuple** | ✅ | ❌ | ✅ | ❌ | Readable fixed records |

> *Dict preserves insertion order in Python 3.7+

---
## 🧠 Decision Guide — Which to pick?

```
Need key-value pairs?          → dict
Need unique items only?        → set / frozenset (if immutable)
Data never changes?            → tuple / namedtuple / frozenset
Need readable field names?     → namedtuple
Need LIFO (undo/back)?         → stack (list)
Need FIFO (queue/schedule)?    → deque
Large numbers, memory matters? → array (or numpy)
Default general collection?    → list
```
