# NumPy for Image Processing
This notebook demonstrates basic image processing operations using NumPy.
Images are treated as multi-dimensional arrays where each pixel can be manipulated directly.

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

## Load an Image
We load an image using PIL and convert it into a NumPy array.
Each pixel is represented as a numerical value.

```python
from PIL import Image

image = Image.open('dog.jpg')
image_array = np.array(image)
plt.imshow(image_array)
plt.axis('off')
```

```python
print("Original Image Array:", image_array)
```

## Image Shape
The shape of the image array tells us height, width, and color channels (RGB).

```python
image_array.shape
```

## Cropping
We can crop an image using NumPy slicing.
This selects a portion of the array.

```python
cropped = image_array[200:800, 150:450]
plt.imshow(cropped)
plt.axis('off')
```

## Flipping
We can flip the image vertically or horizontally using NumPy functions.

```python
flipped = np.flipud(image_array)
plt.imshow(flipped)
plt.axis('off')
```

## Rotation
Rotate the image using NumPy's rot90 function.

```python
rotated = np.rot90(image_array)
plt.imshow(rotated)
plt.axis('off')
```

## Pixel Manipulation
We can modify pixel values directly.
For example, increase brightness by adding a constant value.

```python
bright = np.clip(image_array + 50, 0, 255)
plt.imshow(bright.astype(np.uint8))
plt.axis('off')
```

```python
# Negative of an Image

# Load the image using PIL (Python Imaging Library)
img = Image.open('cat.jpg')

# Convert the image to a NumPy array
img_array = np.array(img)

# Check if the image is grayscale or RGB
is_grayscale = len(img_array.shape) < 3

# Function to create negative of an image
def create_negative(image):
    if is_grayscale:
        # For grayscale images
        negative_image = 255 - image
    else:
        # For color images (RGB)
        negative_image = 255 - image
    return negative_image

# Create negative of the image
negative_img = create_negative(img_array)

# Display the original and negative images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_array)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(negative_img)
plt.title('Negative Image')
plt.axis('off')

plt.tight_layout()
plt.show() 
```
